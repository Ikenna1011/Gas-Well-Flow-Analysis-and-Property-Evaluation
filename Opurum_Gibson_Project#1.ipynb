```python
# All libraries you need for this project. Run this cell once.
import math
import pandas as pd
import matplotlib.pyplot as plt
```

### Step 1: Define function TPC1

```python
# gamma_G is the agrument of of TPC1 function.
# Once the function is called, it returns the value of pseudo critical tempertaure.
def TPC1(gamma_G):
    value_TCP1 = 169.2 + 349.5 * gamma_G - 74 * (gamma_G * gamma_G)
    return value_TCP1
```

##### Test TPC1

```python
# How to call this function, for example for specific gravity of 0.65
gamma_G = 0.65
pseudo_critical_tempertaure = TPC1(gamma_G)
print("The value of pseudo critical tempertaure is: ", pseudo_critical_tempertaure, "\N{SUPERSCRIPT zero}R")
```

##### Plot TPC1

```python
# For plotting Tpc vs. Specific Gravity, we create two lists for x and y. x is gamma_G and Y TPc . 
# gamma_G varies from 0.58 to 1.67 with the increment of 0.05 (you may choose any other incerements).
# We call TPC1 function in a for-loop or while-loop which returns Tpc values,and we append the values to the y_Tpc list by         For writing a while loop, we first define the initial value of Pressure...
#  y_Tpc.append(TPC1_value). We also append gamma_G to the x_gamma_G list. 
y_Tpc=list()
x_gamma_G=list()
gamma_G=0.58
while gamma_G<=1.67:
    TPC1_value=TPC1(gamma_G)
    y_Tpc.append(TPC1_value)
    x_gamma_G.append(gamma_G)
    gamma_G=gamma_G+0.05

print("gamma_G: ",x_gamma_G)
print("Tpc: ",y_Tpc)

# plot: we create a data frame using Panda library for plotting. 
# First we create a dictionary with 
# key (Tpc) and value (y_Tpc) 
# key (gamma_G) and value (x_gamma_G).  
#
data = {
      "gamma_G": x_gamma_G,
      "Tpc":y_Tpc
    }

df = pd.DataFrame(data)
print(df)

df.plot(kind='scatter',x='gamma_G',y='Tpc')
plt.show()
```

### Step 2: Define function PPC1

```python
def PPC1(gamma_G):
    value_PPC1 = 756.8 - 131 * gamma_G - 3.6 * (gamma_G * gamma_G)
    return value_PPC1
```

##### Test PPC1 function

```python
## Running this cell to print PPC1 at 0.65 specific gravity after implementing the above function.

gamma_G = 0.65
pseudo_critical_pressure = PPC1(gamma_G)
print("The value of pseudo critical pressure is: ", pseudo_critical_pressure, "psia")
```

##### Plot PPC1

```python
# Plotting Ppc(gamma_G)

y_Ppc=list()
x_gamma_G=list()
gamma_G=0.58
while gamma_G<=1.67:
    PPC1_value=PPC1(gamma_G)
    y_Ppc.append(PPC1_value)
    x_gamma_G.append(gamma_G)
    gamma_G=gamma_G+0.05

print("gamma_G: ",x_gamma_G)
print("Ppc: ",y_Ppc)

# plot: 

data = {
      "gamma_G": x_gamma_G,
      "Ppc":y_Ppc
    }

df = pd.DataFrame(data)
print(df)

df.plot(kind='scatter',x='gamma_G',y='Ppc')
plt.show()
```
